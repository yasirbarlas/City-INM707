This notebook produces plots from the output "progress.csv" results file when running a Ray RLlib program. The plots and results files are specific to our work on using Proximal Policy Optimisation on the multi-agent PettingZoo Atari 'Space Invaders' environment.

### Import Libraries and Results

We only require Matplotlib for plotting and Pandas for importing .csv files.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_normal = pd.read_csv("ppo_normal.csv")
df_no_kl = pd.read_csv("ppo_no_kl.csv")
df_attention = pd.read_csv("ppo_attention.csv")
```

### Plot Results

We create plots of the mean episode reward during training, the total number of timesteps for these episodes, the minimum reward, and lastly the maximum reward during the episodes.

```python
plt.figure(figsize = (40, 6))
plt.subplot(131)
plt.plot(df_normal["num_env_steps_trained"], df_normal["episode_reward_mean"], label = "KL Loss")
plt.plot(df_attention["num_env_steps_trained"], df_attention["episode_reward_mean"], label = "KL Loss With Attention")
plt.plot(df_no_kl["num_env_steps_trained"], df_no_kl["episode_reward_mean"], label = "No KL Loss")
plt.ylim(600, 1000)
plt.title("PPO: Average Reward Per Episode")
plt.xlabel("Environment Steps")
plt.ylabel("Average Reward")
plt.legend(loc = "upper right")

plt.subplot(132)
plt.plot(df_normal["num_env_steps_trained"], df_normal["sampler_results/episodes_timesteps_total"], label = "KL Loss")
plt.plot(df_attention["num_env_steps_trained"], df_attention["sampler_results/episodes_timesteps_total"], label = "KL Loss With Attention")
plt.plot(df_no_kl["num_env_steps_trained"], df_no_kl["sampler_results/episodes_timesteps_total"], label = "No KL Loss")
plt.ylim(0, 250000)
plt.title("PPO: Time Steps Per Episode")
plt.xlabel("Environment Steps")
plt.ylabel("Episode Time Steps")
plt.legend(loc = "upper right")
plt.savefig("ppo_average_timesteps.pdf")
plt.show()
```

```python
plt.figure(figsize = (40, 6))
plt.subplot(131)
plt.plot(df_normal["num_env_steps_trained"], df_normal["episode_reward_min"], label = "KL Loss")
plt.plot(df_attention["num_env_steps_trained"], df_attention["episode_reward_min"], label = "KL Loss With Attention")
plt.plot(df_no_kl["num_env_steps_trained"], df_no_kl["episode_reward_min"], label = "No KL Loss")
plt.title("PPO: Minimum Reward Per Episode")
plt.xlabel("Environment Steps")
plt.ylabel("Minimum Reward")
plt.legend(loc = "upper right")

plt.subplot(132)
plt.plot(df_normal["num_env_steps_trained"], df_normal["episode_reward_max"], label = "KL Loss")
plt.plot(df_attention["num_env_steps_trained"], df_attention["episode_reward_max"], label = "KL Loss With Attention")
plt.plot(df_no_kl["num_env_steps_trained"], df_no_kl["episode_reward_max"], label = "No KL Loss")
plt.ylim(600, 2200)
plt.title("PPO: Maximum Reward Per Episode")
plt.xlabel("Environment Steps")
plt.ylabel("Maximum Reward")
plt.legend(loc = "upper right")
plt.savefig("ppo_minimum_maximum.pdf")
plt.show()
```
